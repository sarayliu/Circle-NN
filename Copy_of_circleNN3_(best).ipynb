```
import torch
import random
```

```
inputExp = 'x*x+y*y>1'
ins, expected = [], []
for iteration in range(10000):
    x, y = random.uniform(-1.5, 1.5), random.uniform(-1.5, 1.5)
    ins.append([x, y])
    if eval(inputExp):
        expected.append([1.0])
    else:
        expected.append([0.0])
ins = torch.tensor(ins)
expected = torch.tensor(expected)
nodeCts = [2, 15, 5, 1]
print(ins)
print(expected)
print(nodeCts)
```

```
netSpec = [torch.nn.Sigmoid() if i%2 else
           torch.nn.Linear(nodeCts[i//2], nodeCts[1+i//2],bias=i<1)
             for i in range(2*len(nodeCts)-2)]
mynn = torch.nn.Sequential(*netSpec, torch.nn.Linear(1,1,bias=False))
criterion = torch.nn.MSELoss()
print(netSpec)
print(mynn)
print(criterion)
```

```
optimizer = torch.optim.SGD(mynn.parameters(), lr=0.5)
# for epoch in range(2000000+1):
for epoch in range(1155500+1):
  y_pred = mynn(ins)  # Forward propagation
#   print(y_pred)
#   for y_val in range(len(y_pred)):
#     y_pred[y_val] = round(y_pred[y_val].item())
#   print(y_pred)
  loss = criterion(y_pred, expected)  # Compute and print error
  if not epoch%500 or epoch<10:
    print('epoch: ', epoch, ' loss: ', loss.item())

  optimizer.zero_grad()  # Zero the gradients
  loss.backward()  # Back propagation
  optimizer.step()  # Update the weights
```

```
for k,v in mynn.state_dict().items(): 
  print(k,v)
weightList = []
for k,v in mynn.state_dict().items():
  print(k)
  if k != '0.bias':
    wghtList = []
    for val in v:
      for t_val in val:
        print(str(t_val.item()) + ',')
        wghtList.append(t_val.item()) 
    weightList.append(wghtList)
  else:
    biasCt = 2
    for val in v:
      print(val.item())
      weightList[0].insert(biasCt, val.item())
      biasCt += 3
print(weightList)
```

```
testIns, testExpected = [], []
for iteration in range(100000):
    x, y = random.uniform(-1.5, 1.5), random.uniform(-1.5, 1.5)
    testIns.append([x, y])
    if eval(inputExp):
        testExpected.append([1.0])
    else:
        testExpected.append([0.0])
testIns = torch.tensor(testIns)
testExpected = torch.tensor(testExpected)
print(testIns)
print(testExpected)
```

```
test_y_pred = mynn(testIns)
rounded_vals = []
for test_y_val in range(len(test_y_pred)):
    rounded_vals.append(round(test_y_pred[test_y_val].item()))
rounded_vals = torch.tensor(rounded_vals)
failureCt = 0
for iteration in range(100000):
    if testExpected[iteration].item() != rounded_vals[iteration].item():
      failureCt += 1
print(failureCt)
```
